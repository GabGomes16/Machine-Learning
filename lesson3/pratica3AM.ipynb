```python
from ISLP import load_data
# Carregar os dados
df = load_data('Default')
# Preparação dos dados
# Converter a variável 'default' para binária (1 para Yes, 0 para No)
df['default'] = df['default'].apply(lambda x: 1 if x == 'Yes' else 0)
# Converter a variável 'student' para binária
df['student'] = df['student'].apply(lambda x: 1 if x == 'Yes' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split

# Dividir os dados
X = df[['balance', 'income', 'student']]
y = df['default']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
# Modelo com a variavel ’balance’ como preditora
log_reg_balance = LogisticRegression()
log_reg_balance.fit(X_train[['balance']], y_train)
# Modelo com a variavel ’income’ como preditora
log_reg_income = LogisticRegression()
log_reg_income.fit(X_train[['income']], y_train)
# Modelo com a variavel ’student’ como preditora
log_reg_student = LogisticRegression()
log_reg_student.fit(X_train[['student']], y_train)
# Modelo com todas as variaveis preditoras
log_reg_multi = LogisticRegression()
log_reg_multi.fit(X_train, y_train)
```

```python
# Exibir coeficientes
print("Coeficiente de balance:", log_reg_balance.coef_)
print("Coeficiente de income:", log_reg_income.coef_)
print("Coeficiente de student:", log_reg_student.coef_)
print("Coeficientes para balance, income, student:", log_reg_multi.coef_)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix
# Previs~oes e m´etricas para todas as vari´aveis juntas
y_pred_multi = log_reg_multi.predict(X_test)
print("Acuracia:", accuracy_score(y_test, y_pred_multi))
print("Matriz de Confusao:\n", confusion_matrix(y_test, y_pred_multi))
```

```python
import numpy as np
from sklearn.metrics import recall_score, precision_score
# Ajuste do limiar
y_pred_proba = log_reg_multi.predict_proba(X_test)[:,1]
y_pred_adjusted = np.where(y_pred_proba > 0.3, 1, 0)
print("Acuracia com limiar ajustado:", accuracy_score(y_test, y_pred_adjusted))
print("Revocacao com limiar ajustado:", recall_score(y_test, y_pred_adjusted))
print("Precisao com limiar ajustado:", precision_score(y_test, y_pred_adjusted))
```
