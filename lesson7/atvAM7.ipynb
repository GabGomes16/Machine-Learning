```python
# Importando as bibliotecas necessárias
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, LeaveOneOut
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
from sklearn.utils import resample
import seaborn as sns

# Carregando o dataset Titanic
titanic = sns.load_dataset('titanic')

# Pré-processamento de dados
# Remover linhas com valores nulos nas colunas selecionadas
titanic = titanic.dropna(subset=['age', 'embarked', 'sex', 'class', 'fare'])

# Selecionando as features e o target
X = titanic[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = titanic['survived']

# Transformar variáveis categóricas em numéricas
X = pd.get_dummies(X, drop_first=True)

# Padronizando os dados
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Criando o modelo
model = LogisticRegression()

# Função para Hold-Out
def hold_out_evaluation(X, y, model):
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3, stratify=y, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    return precision, recall, f1

# Função para Leave-One-Out
def loo_evaluation(X, y, model):
    loo = LeaveOneOut()
    f1_scores_loo = []
    for train_index, test_index in loo.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        f1_scores_loo.append(f1_score(y_test, y_pred, zero_division=0))
    return np.mean(f1_scores_loo)

# Função para K-Fold
def kfold_evaluation(X, y, model, splits=5):
    skf = StratifiedKFold(n_splits=splits, random_state=42, shuffle=True)
    f1_scores_kfold = cross_val_score(model, X, y, cv=skf, scoring='f1')
    return np.mean(f1_scores_kfold)

# Executando avaliações
hold_out_results = hold_out_evaluation(X_scaled, y, model)
loo_results = loo_evaluation(X_scaled, y, model)
kfold_results_5 = kfold_evaluation(X_scaled, y, model, splits=5)
kfold_results_10 = kfold_evaluation(X_scaled, y, model, splits=10)

hold_out_results, loo_results, kfold_results_5, kfold_results_10
```
